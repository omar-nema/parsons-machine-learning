```python
# GET A BATCH OF CHANNELS FROM ARENA

import requests as req
   
# api-endpoint
URL = "http://api.are.na/v2/channels"

p = {
    "total_pages": 1,
    "per": 20000
}

r = req.get(url = URL, params = p)
channels  = r.json().get('channels')

len(channels)
```

```python
# FOR EACH CHANNEL, GET THE RELEVANT BLOCKS. ONLY SAVE BLOCKS THAT HAVE IMAGE URLS.

channelUrl = "http://api.are.na/v2/channels/"
channel_slugs = []
blocks_raw = []
blocks_metadata = []
blocks_export = {}
requests = []

print('start')
for i, c in enumerate(channels):
    slug = c.get('slug')
    if i % 10 == 0:
        print(str(i) + ' ' + slug)
    URL = channelUrl + slug
    block_data = req.get(URL, {}).json().get('contents', {})
    if block_data:
        block_data = block_data[0]
        block_data = {key: {} if value is None else value for (key, value) in block_data.items()}
        blocks_raw.append(block_data)
         #only add those that have thumb url
        blockObj = {
             'id': block_data.get('id', {}),
            'title': block_data.get('title', {}), 
            'created_at': block_data.get('created_at', {}),    
            'thumb_url': block_data.get('image', {}).get('thumb', {}).get('url', {})
        }
        if block_data.get('id', {}):
            blocks_export[block_data.get('id', {})] = blockObj
        #blocks_metadata.append(blockObj)
        
print('done')
```

```python
# GET IMAGES FROM URL, AND SAVE LOCALLY

from PIL import Image
import matplotlib.pyplot as plt

block_images = []
numFailed = 0
i = 0
for key, value in blocks_export.items():
    if i % 10 == 0:
        print(i)
    i = i +1
    img_meta = value
    img_url = img_meta.get('thumb_url')
    if (img_url):
        img_id = img_meta.get('id')
        file_ext = ".\\Desktop\\arenaui\\images\\" + str(img_meta.get('id')) + '.jpg'
        try:
            img = Image.open(req.get(img_url, stream=True).raw)
            img.convert('RGB').save(file_ext)
            block_images.append(img) 
        except:
            numFailed += 1
            print('exception')
            print(img_url)
 
print('failed', numFailed)
print('success', len(blocks_export))
```

```python
import json

with open('./Desktop/arenaui/imageDir.json', 'w') as outfile:
    json.dump(blocks_export, outfile)
print('done')
```
